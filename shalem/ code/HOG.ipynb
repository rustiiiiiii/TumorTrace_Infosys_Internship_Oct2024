```python
import cv2
import matplotlib.pyplot as plt
from skimage.feature import hog
from skimage import exposure

# Step 1: Read the image
image = cv2.imread('/Users/shalem/Documents/tumourtrace/clasification-roi/test/Benign/BreaDM-Be-1810/SUB1/p-030.jpg', cv2.IMREAD_GRAYSCALE)

# Step 2: Resize the image to 224x224
image_resized = cv2.resize(image, (224, 224))

# Step 3: Plot the resized image and basic histogram side by side
def plot_image_and_histogram(image):
    pixel_values = image.flatten()
    
    plt.figure(figsize=(12, 6))
    
    # Plot the resized image
    plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
    plt.imshow(image, cmap='gray')
    plt.title('Resized Image (224x224)')
    plt.axis('off')  # Hide axis for the image
    
    # Plot the histogram of pixel values
    plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
    plt.hist(pixel_values, bins=256, range=(0, 256), color='gray', alpha=0.7)
    plt.title('Histogram of Pixel Values')
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')
    
    # Show the plots
    plt.tight_layout()
    plt.show()

# Call the function to plot both the image and histogram
plot_image_and_histogram(image_resized)
```
